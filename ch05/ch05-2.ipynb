```python
from google.colab import drive
drive.mount('/content/drive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/drive/MyDrive/ColabIP/images/lena256.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gimg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg.shape)
cv2_imshow(gimg)
```

```python
from random import randint
```

```python
# 10% noise
noiseRatio = 20
whiteMax = noiseRatio/2
blackMax = noiseRatio

```

```python
noiseimg = gimg.copy()

for h in range(noiseimg.shape[0]):
  for w in range(noiseimg.shape[1]):
    val = randint(0, 100)
    if(val < whiteMax):
      noiseimg[h,w] = 255
    elif(val < blackMax):
      noiseimg[h,w] = 0

cv2_imshow(noiseimg)
```

```python
outimg = cv2.medianBlur(noiseimg, 5)
cv2_imshow(outimg)
```
