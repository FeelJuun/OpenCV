```python
from google.colab import drive
drive.mount('/content/drive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/drive/MyDrive/ColabIP/images/1118.png')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gimg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg.shape)
cv2_imshow(gimg)
```

```python
kernel = np.ones((5,1), np.uint8)
dilateimg = cv2.dilate(gimg, kernel, iterations=1)
cv2_imshow(dilateimg)
```

```python
kernel = np.ones((4,1), np.uint8)
erodeimg = cv2.erode(dilateimg, kernel, iterations=1)
cv2_imshow(erodeimg)
```

```python
img = cv2.imread('/content/drive/MyDrive/ColabIP/images/fce(salt_pepper noise).bmp')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gimg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg.shape)
cv2_imshow(gimg)
```
