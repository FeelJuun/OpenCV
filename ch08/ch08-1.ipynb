```python
from google.colab import drive
drive.mount('/content/drive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/drive/MyDrive/ColabIP/images/char.bmp')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gimg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg.shape)
cv2_imshow(gimg)
```

```python
kernel = np.ones((3,3), np.uint8)
dilateimg = cv2.dilate(gimg, kernel, iterations=3)
cv2_imshow(dilateimg)
```

```python
edgeimg = dilateimg - gimg
cv2_imshow(edgeimg)
```

```python
img = cv2.imread('/content/drive/MyDrive/ColabIP/images/milkdrop.bmp')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gimg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg.shape)
cv2_imshow(gimg)
```

```python
gBin = np.zeros((gimg.shape[0], gimg.shape[1]), dtype=np.ubyte)

threshold = 190

for h in range(gimg.shape[0]):
  for w in range(gimg.shape[1]):
    if gimg[h,w] > threshold:
      gBin[h,w] = 255

cv2_imshow(gBin)
```

```python
kernel = np.ones((3,3), np.uint8)
erodeimg = cv2.erode(gBin, kernel, iterations=3)
cv2_imshow(erodeimg)
```

```python
kernel = np.ones((3,3), np.uint8)
dilateimg = cv2.dilate(gBin, kernel, iterations=5)
cv2_imshow(dilateimg)
```
