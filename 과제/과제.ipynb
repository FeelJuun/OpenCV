```python
from google.colab import drive
drive.mount('/content/drive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
gimg = '/content/drive/MyDrive/ColabIP/images/photo.png'
gimg = cv2.imread(gimg, cv2.IMREAD_COLOR)

gray_img = cv2.cvtColor(gimg, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

```python
#히스토그램 평활화를 이용하여 명암과 대비를 개선함.
equalized_img = cv2.equalizeHist(gray_img)

cv2_imshow(equalized_img)
```

```python
#중간값필터링을 이용하여 노이즈들을 제거
median_filtered_img = cv2.medianBlur(equalized_img, 3) # 커널크기 3

cv2_imshow(median_filtered_img)
```

```python
#중간값필터링을 이용하여 노이즈들을 제거
median_filtered_img = cv2.medianBlur(equalized_img, 5) # 커널크기 5

cv2_imshow(median_filtered_img)
```

```python
#중간값필터링을 이용하여 노이즈들을 제거
median_filtered_img = cv2.medianBlur(equalized_img, 7) # 커널크기 7

cv2_imshow(median_filtered_img)
# 중간값필터의 커널크기를 7까지 늘렸을땐 노이즈들이 많이 사라졌지만 이미지가 흐릿해짐.
```
