```python
from google.colab import drive
drive.mount('/content/drive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/drive/MyDrive/ColabIP/images/noise_dog.png')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gimg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg.shape)
cv2_imshow(gimg)
```

중간값 필터링에서 커널크기가 3일경우 노이즈가 남아있으므로, 커널크기 5를 사용

```python
median_filtered_img = cv2.medianBlur(gimg, 3) # 커널크기 3

cv2_imshow(median_filtered_img)
```

```python
median_filtered_img = cv2.medianBlur(gimg, 5) # 커널크기 5

cv2_imshow(median_filtered_img)
```

중간값 필터링이 완료된 이미지를 화질을 개선하기위해 히스토그램 평활화 사용

```python
equalized_img = cv2.equalizeHist(median_filtered_img)

cv2_imshow(equalized_img)
```
